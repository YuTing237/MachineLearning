```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split   #sklearn包含資料集、演算法、資料處理…等模組。
from sklearn import tree #從sklearn導入決策樹

# 讀入鳶尾花的資料
iris = load_iris()
iris_data = iris.data
iris_label = iris.target

# 切分訓練與測試資料
train_data, test_data, train_label, test_label = train_test_split(iris_data, iris_label, test_size = 0.3)  
#test_size以訓練資料比例70%，測試資料比例30%。

clf = tree.DecisionTreeClassifier()
clf.fit(train_data, train_label)

print(clf.score(train_data,train_label)) # 印出訓練資料正確率
print(clf.score(test_data, test_label)) # 印出測試資料正確率
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("iris") #將結果存成pdf檔案
```

```python
import pydotplus  
from IPython.display import Image  #從IPython.display導入圖像 
dot_data = tree.export_graphviz(clf, out_file=None, 
                feature_names=iris.feature_names, 
                #特徵名稱等於iris.feature_names
                class_names=iris.target_names, 
                #類別名稱等於iris.target_names 
                filled=True, rounded=True,  
                special_characters=True)
                #特殊字符等於真  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  #將結果用圖形化呈現
```
